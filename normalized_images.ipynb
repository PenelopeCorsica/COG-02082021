```python
im1 = cv2.imread('1.jpg')
im2 = cv2.imread('5.jpg')

im1_resized = cv2.resize(im1, (100,100))
im2_resized = cv2.resize(im2, (100,100))

# A and B -> xA + yB -> x and y are weights! 
weightedSum = cv2.addWeighted(im1_resized, 0.75, im2_resized,0.75, 0)
# 0.5 * im1 + 0.4 * im2 = y
# Only for google colab, for rest use cv2.imshow(weightedSum)
cv2_imshow(weightedSum)
```

```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, array_to_img, load_img

generator = ImageDataGenerator(rotation_range=40, shear_range=0.25, zoom_range=0.25,
                               horizontal_flip=True, brightness_range=(0.5, 1.5))
# brightness will be adjusted from 50% to 150% 
img = load_img('5.jpg') # if there was no COMPUTER vision work, avoid CV2 
# in production , if you just had to load images, no need of CV2 library! 
img = img_to_array(img)
img[1]
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(img)
```

```python
img_as_sequence_element = img.reshape((1,) + img.shape)
# the same output as np.expand_dims(img, axis=0)
counter = 0
for batch in generator.flow(img_as_sequence_element, batch_size=1, save_to_dir='augs',
                            save_prefix='panda', save_format='jpeg'):
  if counter == 30:
    break
  counter = counter + 1
```
